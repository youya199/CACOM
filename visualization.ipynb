```python
# Import packages
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

def visual(list_of_angles):
    # Create figure and add axes
    fig = plt.figure(figsize=(6, 4))
    ax = fig.add_subplot(111)
    ax.set_xlabel('time')
    ax.set_ylabel('angle')

    # time values
    epo=len(list_of_angles)
    T = np.linspace(0, epo*1/240, epo)
    # angle values
    A=list_of_angles

    # Create  plot
    ax.plot(T, A, linewidth=2.5)
```

```python
import math
def softmax(T,x):
    l=len(x)
    sum=0
    for i in range(l):
        sum+=math.exp(x[i]/T)
    q=[]
    for i in range(l):
        q.append(math.exp(x[i]/T)/sum)
    return q
```

```python
x=[1,8,7,2,3,5,9,3,3,2]
for j in range(5):

    T=j+2
    q=softmax(T,x)
    visual(q)
```
